# Q11)  Suppose we want to estimate the average weight of an adult male in    Mexico. We draw a random sample of 2,000 men from a population of 3,000,000 men and weigh them. We find that the average person in our sample weighs 200 pounds, and the standard deviation of the sample is 30 pounds. Calculate 94%,98%,96% confidence interval?

```python
import scipy.stats as st
```

```python
conf_int_94 = st.t.interval(alpha=0.94,df=1999,loc=200,scale=30/np.sqrt(2000))
```

```python
conf_int_98 = st.t.interval(alpha=0.98,df=1999,loc=200,scale=30/np.sqrt(2000))
```

```python
conf_int_96 = st.t.interval(alpha=0.96,df=1999,loc=200,scale=30/np.sqrt(2000))
```

```python
print("94 % confidence interval= ",conf_int_94)
print("98 % confidence interval= ",conf_int_98)
print("96 % confidence interval= ",conf_int_96)
```

# Q 24)   A Government  company claims that an average light bulb lasts 270 days. A researcher randomly selects 18 bulbs for testing. The sampled bulbs last an average of 260 days, with a standard deviation of 90 days. If the CEO's claim were true, what is the probability that 18 randomly selected bulbs would have an average life of no more than 260 days Hint:  rcode = pt(tscore,df)  ,df =degrees of freedom


```python
pop_mean= 270
sample = 18
sample_mean=260
sample_std=90
import math
```

```python
t_score = ((sample_mean-pop_mean)*math.sqrt(sample)/(sample_std))
t_score
```

```python
prob=st.t.cdf(t_score,df=17)
```

```python
print("probability of bulbs with life of no more than 260 days",prob.round(3)*100)
```
