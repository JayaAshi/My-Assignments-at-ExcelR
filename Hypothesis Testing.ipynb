# 1.A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from scipy import stats
import statsmodels.api as sm
```

```python
labels = ['Unit A', 'Unit B']
sb.distplot(data['Unit A'])
sb.distplot(data['Unit B'])
plt.legend(labels)
```

```python
# H0 = diameter of the cutlet between two units are same
# H1 = diameter of the cutlet between two units are not same
```

```python
UnitA=data['Unit A']
UnitA
```

```python
UnitB=data['Unit B']
UnitB
```

```python
# 2  Tailed Ttest
#find pvalue
```

```python
pvalue= stats.ttest_ind(UnitA,UnitB)
pvalue
```

```python
#significance level,alpha/2 = 0.025
#pvalue>alpha/2 hence fail to reject Null hypothesis
#There is no difference b/w both diameters
```

# 3.Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.


```python
data2= pd.read_csv(r'C:\Users\Shiva\Desktop\ExcelR assignments\Hypothesis Testing\BuyerRatio (1).csv')
data2
```

```python
##  H0 = All proportions are equal
## H1 = Not all proportions are equal
## significance level , alpha = 5%
```

```python
observed= [[50,142,131,70],
        [435,1523,1356,750]]
```

```python
stats.chi2_contingency(observed)
```

```python
observed= np.array([50,142,131,70, 435,1523,1356,750])
expected=np.array([  42.76531299,  146.81287862,  131.11756787,   72.30424052,442.23468701, 1518.18712138, 
                   1355.88243213,  747.69575948])
```

```python
stats.chisquare(observed,expected,ddof=3)
```

```python
## pvalue>significance 0.05 alpha 
## fail to reject the null hypothesis
## there is no similarity b/w male and female buyers
```
